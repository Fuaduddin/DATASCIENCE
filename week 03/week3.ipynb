```python
# this function takes particular values in a particular row of data and finds min and max values
# the value returns a new row of data and return a new row data.
import numpy as np
def min_max(row):
    data = row[['POPESTIMATE2010', # papoluation estimate
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    return pd.Series({'min': np.min(data), 'max': np.max(data)}) # return new series
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv('census.csv')
df = df[df['SUMLEV']==50] # calling only sumlev == 50
df
```

```python
%%timeit -n 10 # koto shomoy dhorae ai cel cholbae
for state in df['STNAME'].unique(): # to find unique value
    avg = np.average(df.where(df['STNAME']==state).dropna()['CENSUS2010POP']) # calculating average using numpy
    print('Counties in state ' + state + ' have an average population of ' + str(avg)) # printing the value
```

```python
%%timeit -n 10
# grouping by state name
for group, frame in df.groupby('STNAME'): # here we want to group by state name
    avg = np.average(frame['CENSUS2010POP']) # calculating average of cencus
    print('Counties in state ' + group + ' have an average population of ' + str(avg)) # print
```

# Pivot Tables

```python
import pandas as pd
import numpy as np
#http://open.canada.ca/data/en/dataset/98f1a129-f628-4ce4-b24d-6f16bf24dd64
df = pd.read_csv('cars.csv')
```

```python
df.head()
```

```python
def create_category(kw):
    if(kw>=1) & (kw<=100):
        return "First tier tor car"
    elif (kw>=100) & (kw<=150):
        return "Second tier tor cars"
    elif(kw>=151) & (kw<=250):
        return "third tier top cars"
```

```python
df.pivot_table(values='(kW)', index='YEAR', columns='Make', aggfunc=np.max) # average rating
```

```python
df.pivot_table(values='(kW)', index='YEAR', columns='Make', aggfunc=[np.mean,np.min], margins=True)
# max rating
```

```python
df.pivot_table(values='(kW)',index='YEAR',columns="Make",aggfunc=[np.mean,np.max],margins=True).head()
# list of different function
# hierarchical column names
```

```python
df.pivot_table(values='(kW)',index='YEAR',columns="Make",aggfunc=[np.mean,np.max],margins=True).head()
```
